```python
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": round((df.isnull().sum()/len(df))*100,2)
})
missing.sort_values(by="Percentage", ascending=False)
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.nunique().sort_values(ascending=False)
```

```python
categorical_columns = df.select_dtypes(include="object").columns
for col in categorical_columns:
    print("="*50)
    print(col)
    print(df[col].unique())
```

```python
num_cols = df.select_dtypes(include=np.number).columns
num_cols
```

```python
cat_cols = df.select_dtypes(include="object").columns
cat_cols
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
print("Date Conversion Successful")
```

```python
print("Minimum Date:", df["Order Date"].min())
print("Maximum Date:", df["Order Date"].max())
```

```python
for col in cat_cols:
    print("\n","="*60)
    print(col)
    print(df[col].value_counts().head(10))
```

```python
df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python

print("Total Orders:", len(df))
print("Total Customers:", df["Customer ID"].nunique())
print("Total Products:", df["Product Name"].nunique())
print("Total Categories:", df["Category"].nunique())
print("Total States:", df["State"].nunique())
print("Total Cities:", df["City"].nunique())
```
